```python
# weekly update (January 31, 2022)
all_order_and_referring_physicians = dd.read_csv("../data/Order_and_Referring_01_31_2022.csv", blocksize="10GB")
```

```python
all_order_and_referring_physicians["NPI"] = all_order_and_referring_physicians["NPI"].astype(int)
cms_nppes_df["NPI"] = cms_nppes_df["NPI"].astype(int)
```

```python
joined = cms_nppes_df.merge(all_order_and_referring_physicians, left_on="NPI", right_on="NPI", how="inner")
```

```python
intersection_df = joined.compute()
```

```python
intersection_df["Last Update Date"] =  pd.to_datetime(
    intersection_df["Last Update Date"], errors='coerce').dt.date
```

```python
intersection_df["Last Update Date"]  = pd.to_datetime(intersection_df["Last Update Date"])
```

```python
intersection_df["NPI Deactivation Date"]  = pd.to_datetime(intersection_df["NPI Deactivation Date"])
intersection_df["NPI Reactivation Date"]  = pd.to_datetime(intersection_df["NPI Reactivation Date"])
```

```python
intersection_df["Last Update Date (year)"] = intersection_df["Last Update Date"].dt.year

intersection_df["Last Update Date (month)"] = intersection_df["Last Update Date"].dt.month
```

```python
intersection_df["NPI Deactivation Date (year)"] = intersection_df["NPI Deactivation Date"].dt.year

intersection_df["NPI Deactivation Date (month)"] = intersection_df["NPI Deactivation Date"].dt.month
```

```python
update_years_df = intersection_df.groupby("Last Update Date (year)").nunique()["NPI"]

update_years_df = update_years_df.reset_index()

update_years_df["Last Update Date (year)"] = np.round(update_years_df["Last Update Date (year)"].astype(int))
```

```python

plt.figure(figsize=(10,5))
sns.barplot(update_years_df["Last Update Date (year)"], update_years_df["NPI"].values, alpha=0.8)
plt.title('When do providers tend to update (yearly)?')
plt.ylabel('Number of Providers', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
update_months_df = intersection_df.groupby("Last Update Date (month)").nunique()["NPI"]

update_months_df = update_months_df.reset_index()

update_months_df["Last Update Date (month)"] = np.round(update_months_df["Last Update Date (month)"].astype(int))
```

```python
intersection_df['Provider Enumeration Date']  = pd.to_datetime(intersection_df['Provider Enumeration Date'])
```

```python
intersection_df['Certification Date']  = pd.to_datetime(intersection_df['Certification Date'])
```

```python
plt.figure(figsize=(10,5))
sns.barplot(update_months_df["Last Update Date (month)"], update_months_df["NPI"].values, alpha=0.8)
plt.title('When do providers tend to update (monthly)?')
plt.ylabel('Number of Providers', fontsize=12)
plt.xlabel('Month of Year (1 - 12)', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
intersection_df_dates_df = intersection_df[["NPI",'Certification Date',
                 'Provider Enumeration Date',"NPI Reactivation Date", "NPI Deactivation Date"]]
```

```python
intersection_df_dates_df_valid_dates = intersection_df_dates_df[
                        (~intersection_df_dates_df["NPI Deactivation Date"].isnull()) &
                         (~intersection_df_dates_df["Provider Enumeration Date"].isnull())]
```

```python
intersection_df_dates_df_valid_dates_validation = intersection_df_dates_df[
                        (~intersection_df_dates_df["NPI Deactivation Date"].isnull()) &
                         (~intersection_df_dates_df["NPI Reactivation Date"].isnull())]
```

```python
intersection_df_dates_df_valid_dates_validation.shape
```

```python
intersection_df_dates_df_valid_dates["time_span_days"] = (intersection_df_dates_df_valid_dates[
    "NPI Deactivation Date"] - intersection_df_dates_df_valid_dates[
    "Provider Enumeration Date"]).dt.days
```

```python
plt.figure(figsize=(15,8))
sns.displot(intersection_df_dates_df_valid_dates["time_span_days"])
plt.title('How many days typically do we see between registration and deactivation - according to NPPES?')
plt.ylabel('Frequency (n = 3794)', fontsize=12)
plt.xlabel('Days', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
intersection_df_dates_df_valid_dates_validation["time_span_days_reactivation"] = (
    intersection_df_dates_df_valid_dates_validation["NPI Reactivation Date"] - intersection_df_dates_df_valid_dates_validation[
    "NPI Deactivation Date"]).dt.days
```

```python
plt.figure(figsize=(15,8))
sns.displot(intersection_df_dates_df_valid_dates["time_span_days"])
plt.title('How many days typically do we see between deactivation and reactivation - according to NPPES?')
plt.ylabel('Frequency (n = 3794)', fontsize=12)
plt.xlabel('Days', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```
